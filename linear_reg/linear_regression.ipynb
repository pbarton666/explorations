```python
import random
import warnings

import matplotlib.pyplot as plt

import numpy as np
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import seaborn as sns

import statsmodels.api as sm
import sklearn.datasets as ds

%matplotlib inline
```

## Basics

Let's say that you believe that two concepts might be related.  You don't **know** this, but have good reasons to expect that to be true.    Let's further assume that you believe that the relationship is roughly linear.  That might be Fahrenheit versus Kelvin temperature measurements that tend to rise and fall together.   Or it might be the size of a dog and the frequency of the musical note it hits when it howls.

You might consider using linear reguression to better understand that relationship.

In its most simple form, linear regression is a way to find the best possible fit a straight line through a cloud of data.  It can be used to estimate the slope (y) and intercept (b) of an equation in the form of:

$\Large y = mx + b$

To illustrate, let's make up some data:

```python
x = [random.random()*50 for _ in range(50)]
y = [.5 * x_val -2 + random.random()*5 for x_val in x]
plt.scatter(x,y)
```

## First-order Inspection

It's easy to "eyeball" this and make some observations.    The x and y data are clearly related, right?   The relationship is clearly linear, right?   In fact, you could make a pretty decent guess about the slope and intercept.

Let's try:   slope ~ .5; intercept ~ 0

```python
slope_guess = .5
intercept_guess = 0

x_guess = np.linspace(0, 50, 1000)
y_guess = [ slope_guess * x + intercept_guess for x in x_guess]

plt.scatter(x,y)
plt.plot(x_guess, y_guess, color='r')
```

Our estimate looks pretty good when we plot it out.    But, maybe other values might also fit pretty well.    Let's tweak our slope and intercept a bit and see what we get:

```python
# Our data
plt.scatter(x,y)

# First guess
x_guess = np.linspace(0, 50, 1000)
y_guess = [ slope_guess * x + intercept_guess for x in x_guess]
plt.plot(x_guess, y_guess, color='r')

# Tweak the slope
y_guess = [ (slope_guess -.03) * x + intercept_guess for x in x_guess]
plt.plot(x_guess, y_guess, color='g')

# Tweak the intercept
y_guess = [ slope_guess * x + (intercept_guess + 1) for x in x_guess]
plt.plot(x_guess, y_guess, color='orange')
```

It gets pretty hard to tell which one fits best, doesn't it?  But at least we have a sense for what the range of values that the "right answer" has to be in.

## Applying the regression model

That's where a technique like linear regression will help out.   Linear regression routines can be found in many places in the Python ecosystem.   One of these is found in the "scikit-learn" library (home to many other statistical and machine learning tools).  

We can use it with a three-step process:
- import the LinearRegression class, 
- create an instance, and 
- run the fit() method.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model.fit(np.array(x).reshape(-1,1), np.array(y)) #convert x -> 2D np.ndarray; y -> 1D np.ndarray
```

As with most of the tools in scikit-learn, the model is an instance of an analytical class.   The class has all the machinery required to implement the statistical technique.

Unlike tools such as SPSS and SAS, the results are also stored in the model object and can be extract by one of its internal methods.

Here. we tap the predict() method.  It contains the estimates of the slope and intercept parameters.  It accepts the "independent variable" $x$ and returns the corresponding "dependent variable" $y$
.

```python
x_fitted = np.linspace(0, 50, 1000)
y_fitted = model.predict(x_fitted[:, np.newaxis])
plt.scatter(x, y)
plt.plot(x_fitted, y_fitted, color='y')
```

## Extract model parameters

We can tap into the model object once again to find the actual parameter estimates generated by our analysis.   The slope is stored in an array of model.coef_ parameters.   Because we only had one independent variable, we look at the first element.   By default, we requested the intercetp - that's stored as model.intercept_.

```python
print(f"Slope:  {model.coef_[0]:.3f}  Intercept:  {model.intercept_:.3f}")
```

# How Does Regression Work?

Regression examines relationships between the observed values of a variable of interest - call it a "dependent variable" - and the observed values of other variable(s) - usually called "independent variables".   

The idea is to create an equation that uses a weighted, linear combination of the independent variables to explain as much of the variance of the dependent variable as possible.

That's a mouthful, I know.   It's a little easier in the language of math.   

Using vectors, we have an equation with the dependent variable $Y$ calculated as a function of the independent variables $\mathbb{Y}$, their slopes $\mathbb{X}$, and errors $\mathbb{E}$:


$\mathbb{Y}$ = $\mathbb{X}$ * $\mathbb{B}$ + $\mathbb{E}$

... or in an expanded form like so:

The goal is to minimize the error term - that's the difference between the true values **y** and what we're able to calculate with the best **b** terms we can manage.   In practice, we usually consider the square of the error "least squares"(this keeps things positive).   Graphically, it looks like this:

The estimated relationship between **x** and **y** is shown as the blue line.   The red dots indicate the observed pairs of individual (x, y) values.  

The distance between the observed and estimated values is the error (presumably caused by an unmeasured disturbance).  

You can imagine the estimating line sliding up and down and rotating with the intercept as its pivot until it minimimzes the total distance represented by the green lines.

From a mathematical perspective the error is minimized when the slope and intercept achieve these values:  

(For a complete derivation, please cf.
https://stats.stackexchange.com/questions/46151/how-to-derive-the-least-square-estimator-for-multiple-linear-regression)

 $\Large\mathbb{B}=\frac{\sum_{i=1}^{n} (X_i - \bar{X} ) (Y_i - \bar{Y} )} {\sum_{i=1}^{n} (X_i - \bar{X} )^2}  = Cov(X,Y)  /  Var(X)$ 

 $\Large slope = \bar{Y} - \mathbb{B} * \bar{X}  $

If you look at the right side of the equation for $ \mathbb{B}$, you'll see that another way think about it is a way to distribute the entire covariance of the system over the individual independent variables.

# Simple Multiple Regression 

Here, we build a slightly more complicated model that attempts to predict pulse rate.  We'll use a well-known dataset containing exercise and physiology data from a small group of volunteers.

We'll take a few extra steps including:
- review bivariate relationships
- review assumptions underlying linear regression
- examine a "scorecard" to see how well we've done and check assumptions.

## Linear Regression Assumptions

### Linearity

There's a linear relationship between the actual and predicted values.

### Homoscedasticity

The variance of the residuals is the same across the data set.

### Independence

Observations are independent of one another.

### Normalicy

Residuals are normally distributed.  Also, the independent variables are multivariate-normal.   Implicitly assumes lack of outliers.

## A Real Dataset

For our next model we'll use a well-known dataset avaiable from the scikit-learn library.   It's stored in a special format that includes metadata and a description of the project.

```python
raw_data = ds.load_linnerud()
print(raw_data['DESCR'])
```

## Hypothesis

Let's say we take pulse rate to be a good measure of health - at least cardiovascular health.

The lower one's resting pulse rate, the better.  

Let's also say that we think that chin-ups, sit-ups and jumping jacks accomplished each day are decent indicators of exercise (the more, the better).   

Finally, let's also say that weight and waist size are negatively associated with good health.   Who doesn't feel healthier after losing a bit of weight?

## Acquire Data

```python
# Build a DataFrame with contents of the data set
exercise = pd.DataFrame(raw_data['data'], columns=raw_data['feature_names'])
phys = pd.DataFrame(raw_data['target'], columns=raw_data['target_names'])

# Merge the two bits of it using the row index
data = pd.merge(exercise, phys, right_index=True, left_index=True)
data.head(3)
```

```python
# Build DataFrame objects with contents of the data set
exercise = pd.DataFrame(raw_data['data'], columns=raw_data['feature_names'])
phys = pd.DataFrame(raw_data['target'], columns=raw_data['target_names'])

# Merge on the row index
data = pd.merge(exercise, phys, right_index=True, left_index=True)
data.head(3)
```

```python
# A general method for reporting results.

def report(results, details=False):
    print(f"Regression Results (dependent is {dv_col})\n{'='*40}")
    for name, param, t in zip(results.params.index, results.params, results.tvalues):
        print(f"{name:10}   b-coeff: {param:6.3f}  t: {t:6.3f}")
        
    if details:
        print(results.summary())
```

## Run a Regression

```python
dv_col = 'Pulse'
dv = data[dv_col]
ivs = data[[ 'Weight',  'Chins', 'Situps', 'Jumps' ]]
model = sm.OLS(dv, ivs)
results = model.fit()
report(results)
```

## Interpret First-order Results

This tells us that Weight and Stiups are both significantly related to Pulse.   Roughly, for every pound gained we can predict .23 extra heartbeats per minute.   This fits our expectations.  We expect the exercises to lower resting pulse, but the t-statistics < |2| are a bit suspicious - maybe they are not significant.


## Dig In A Bit


```python
report(results, details=True)
```

## Key Takeaways:

- The model explains a ton of variance around 95% (the r-squared metric)


- At .05 confidence, Weight has the only significant coefficient.  Situps are pretty close, but the sign of the coeffficient isn't what's expected.


- We have to reject the notion that the residuals are normal (Omnibus and Jarque-Bera have high p-values)


Alas and alack, we're back to the drawing board.

## Other Flavors of Regression

Simple regression is the cornerstone of a large family of analytical techniques including:

- Multivariate regression (multiple independent variables)
- Logistic regression (fit a logit curve to a binomial dependent variable)
- Ridge regression (introduce bias to mitigate correlation among independent variables)
- Lasso regression (drive some coefficient values to zero to allow feature selection)
- Polynomial regression (fit data to a non-linear model)
- Quantile regression (regresses against the median or other quantiles of the dependent variable)
- Nonparametric methods (do not use known distributions as "prior" assumptions)

## Interpreting a "Scorecard"

```python
report(results, details=True)
```

What this means:

R-squared:  How much of the variance of the dependent variable is explained.

Adj. R-squared:  Adds a penalty for independent variables (over-specification)

AIC:  Akaike Information Criteria - penalizes over-specification.   exp((AICminAICi)/2) can be interpreted as being proportional to the probability that the ith model minimizes the (estimated) information loss.  Used like a likelihood-ratio test when comparing models.

BIC: Bayesian Information Criteria - like the AIC, based on the likelihood function.  Penalizes over-specification more than AIC.  Low values are bettter.

Log-Likelihood is the log(probability that we would make these ovservations given the assumed probabiliy distribution).

Jarque-Bera - tests if the skewness and kurtosis are consistent with a normal distribution (H0).

Omnibus - tests if the residuals are normal (H0)

Kurtosis - "peakedness" of the distribution (normal = 3)

Skew - a measure of symmetry (normal = 0)

Durbin-Watson - a measure of homoscedacity (errors are consistent)

F-statistic (H0 is "dependent variables have no effect")
